```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.model_selection import LeaveOneOut
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt

df = pd.read_excel('hepatitis.xlsx',  na_values="?")
df
```

```python
# Input
x = df.drop(columns=['#','CLASS'])
x
```

```python
# Change "?" To Dominant in 1 Column

x['Steroid'].fillna(df.groupby('CLASS')['Steroid'].transform(lambda x: pd.Series.mode(x)[0]), inplace = True)
x['Fatique'].fillna(df.groupby('CLASS')['Fatique'].transform(lambda x: pd.Series.mode(x)[0]), inplace = True)
x['Malaise'].fillna(df.groupby('CLASS')['Malaise'].transform(lambda x: pd.Series.mode(x)[0]), inplace = True)
x['Anorexia'].fillna(df.groupby('CLASS')['Anorexia'].transform(lambda x: pd.Series.mode(x)[0]), inplace = True)
x['Liver Big'].fillna(df.groupby('CLASS')['Liver Big'].transform(lambda x: pd.Series.mode(x)[0]), inplace = True)
x['Liver Firm'].fillna(df.groupby('CLASS')['Liver Firm'].transform(lambda x: pd.Series.mode(x)[0]), inplace = True)
x['Spleen Palpable'].fillna(df.groupby('CLASS')['Spleen Palpable'].transform(lambda x: pd.Series.mode(x)[0]), inplace = True)
x['Speiders'].fillna(df.groupby('CLASS')['Speiders'].transform(lambda x: pd.Series.mode(x)[0]), inplace = True)
x['Ascites'].fillna(df.groupby('CLASS')['Ascites'].transform(lambda x: pd.Series.mode(x)[0]), inplace = True)
x['Varices'].fillna(df.groupby('CLASS')['Varices'].transform(lambda x: pd.Series.mode(x)[0]), inplace = True)
x
```

```python
x = x.replace(to_replace = ['yes','no'],value = ['1','0'])
x
```

```python
x["Bilirubin"] = x["Bilirubin"].fillna(df.groupby('CLASS')['Bilirubin'].transform('mean'))
x["Alk Phosphate"] = x["Alk Phosphate"].fillna(df.groupby('CLASS')['Alk Phosphate'].transform('mean'))
x["SGOT"] = x["SGOT"].fillna(df.groupby('CLASS')['SGOT'].transform('mean'))
x["Albumin"] = x["Albumin"].fillna(df.groupby('CLASS')['Albumin'].transform('mean'))
x["Protime"] = x["Protime"].fillna(df.groupby('CLASS')['Protime'].transform('mean'))
x
```

```python
# Output
y = df['CLASS']
y
```

```python
# Input Testing

x2 = df.drop(columns=['#','CLASS'])
x2 = x2.replace(to_replace = ['yes','no'],value = ['1','0'])
x2
```

```python
# Output Testing

y2 = df['CLASS']
y2
```

```python
loo = LeaveOneOut()

true_y, pred_y = list(), list()
true_y, pred_y

i = 0
for train_index, test_index in loo.split(x):
    train_x, test_x = x.iloc[train_index, :], x2.iloc[test_index, :]
    i+=1
    print("==>", i, "<==")
    print(train_x)
    print("train_x", "-+"*50)
    print(test_x)
    print("test_x", "-+"*50)
    
    train_y, test_y = y.iloc[train_index], y2.iloc[test_index]
    print(train_y)
    print("train_y", "-|"*50)
    print(test_y)
    print("test_y", "-|"*50)

    nan_values = test_x.isna()
    print(nan_values)
    print("nan_values", "=="*50)
    nan_columns = nan_values.any()
    print(nan_columns)
    print("nan_columns", "=="*50)
    
    columns_with_nan = test_x.columns[nan_columns].tolist()
    print(columns_with_nan)
    print("columns_with_nan", "xx"*50)
    train_x = train_x.drop(columns_with_nan,axis = 1)
    print(train_x)
    print("train_x Drop NaN", "xx"*50)
    test_x = test_x.dropna(axis='columns')
    print(test_x)
    print("test_x Drop NA", "xx"*50)

    #KNN Classifier
    classifier = KNeighborsClassifier(n_neighbors=3)
    classifier.fit(train_x, train_y)

    # evaluate model
    yhat = classifier.predict(test_x)
    print("===> PREDICT yhat: ", yhat)
    yhatt=yhat.tolist()
    print("===> PREDICT yhatt: ", yhatt)
    test_yy=test_y.tolist()
    print("===> REAL test_yy: ", test_yy)

    # store
    true_y.append(test_yy[0])
    print("===> STORE true_y: ", true_y)
    pred_y.append(yhatt[0])
    print("===> STORE pred_y: ", pred_y)
```

```python
acc = accuracy_score(true_y, pred_y)
acc = acc*100
error = 100-acc
print(pred_y)
print('Accuracy: %.3f' % acc)
print('Accuracy: %.3f' % error)

print(classification_report(true_y, pred_y))
```
