```python
import pandas as pd

# IMPORT DATASET
dataset = pd.read_csv('C:\Hanif Izzudin Rahman D4 EB 2016\S2 Elektro 2020 - PENS\Semester 2\Knowledge Discovery - Ali Ridho\Praktikum\P1_Data Preprocessing_Hepatitis\Assaigment1.csv')
dataset
```

```python
# Input Data
Input = dataset.iloc[:,0:19]
Input
```

```python
Input = Input.replace(to_replace=['no', 'yes'], value=[0, 1])
Input
```

```python
# Output Data
Output = dataset.iloc[:,-1]
Output
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=3)
classifier.fit(Input, Output)
```

```python
Predict = classifier.predict(Input)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(Output, Predict))
print(classification_report(Output, Predict))
```

```python
from sklearn.model_selection import LeaveOneOut
# create loocv procedure
cv = LeaveOneOut()

from sklearn.model_selection import cross_val_score
# evaluate model
scores = cross_val_score(classifier, Input, Output, scoring='accuracy', cv=cv, n_jobs=-1)

from numpy import mean
from numpy import std
# report performance
print('Accuracy: %.3f (%.3f)' % (mean(scores), std(scores)))
```
