```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.model_selection import LeaveOneOut
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

from sklearn.preprocessing import MinMaxScaler, StandardScaler

import matplotlib.pyplot as plt

df = pd.read_excel('hepatitis.xlsx',  na_values="?")
df
```

```python
# Input
x = df.drop(columns=['#','CLASS'])
x
```

```python
# Change "?" To Dominant in 1 Column

x['Steroid'].fillna(df.groupby('CLASS')['Steroid'].transform(lambda x: pd.Series.mode(x)[0]), inplace = True)
x['Fatique'].fillna(df.groupby('CLASS')['Fatique'].transform(lambda x: pd.Series.mode(x)[0]), inplace = True)
x['Malaise'].fillna(df.groupby('CLASS')['Malaise'].transform(lambda x: pd.Series.mode(x)[0]), inplace = True)
x['Anorexia'].fillna(df.groupby('CLASS')['Anorexia'].transform(lambda x: pd.Series.mode(x)[0]), inplace = True)
x['Liver Big'].fillna(df.groupby('CLASS')['Liver Big'].transform(lambda x: pd.Series.mode(x)[0]), inplace = True)
x['Liver Firm'].fillna(df.groupby('CLASS')['Liver Firm'].transform(lambda x: pd.Series.mode(x)[0]), inplace = True)
x['Spleen Palpable'].fillna(df.groupby('CLASS')['Spleen Palpable'].transform(lambda x: pd.Series.mode(x)[0]), inplace = True)
x['Speiders'].fillna(df.groupby('CLASS')['Speiders'].transform(lambda x: pd.Series.mode(x)[0]), inplace = True)
x['Ascites'].fillna(df.groupby('CLASS')['Ascites'].transform(lambda x: pd.Series.mode(x)[0]), inplace = True)
x['Varices'].fillna(df.groupby('CLASS')['Varices'].transform(lambda x: pd.Series.mode(x)[0]), inplace = True)
x
```

```python
x = x.replace(to_replace = ['yes','no'],value = ['1','0'])
x
```

```python
x["Bilirubin"] = x["Bilirubin"].fillna(df.groupby('CLASS')['Bilirubin'].transform('mean'))
x["Alk Phosphate"] = x["Alk Phosphate"].fillna(df.groupby('CLASS')['Alk Phosphate'].transform('mean'))
x["SGOT"] = x["SGOT"].fillna(df.groupby('CLASS')['SGOT'].transform('mean'))
x["Albumin"] = x["Albumin"].fillna(df.groupby('CLASS')['Albumin'].transform('mean'))
x["Protime"] = x["Protime"].fillna(df.groupby('CLASS')['Protime'].transform('mean'))
x
```

```python
# Output
y = df['CLASS']
y
```

```python
# Input Testing

x2 = df.drop(columns=['#','CLASS'])
x2 = x2.replace(to_replace = ['yes','no'],value = ['1','0'])
x2
```

```python
# Output Testing

y2 = df['CLASS']
y2
```

```python
# Minmax "0-1"
print("======> MINMAX <======")

minmax_scaler = MinMaxScaler()

print(x[:1])
print("==> X ", "=="*25)
x_minmax = minmax_scaler.fit_transform(x)
x_minmax = pd.DataFrame(x_minmax)
print(x_minmax.head())

print(x2[:1])
print("==> X2 ", "=="*25)
x2_minmax = minmax_scaler.fit_transform(x2)
x2_minmax = pd.DataFrame(x2_minmax)
print(x2_minmax.head())
```

```python
loo = LeaveOneOut()

true_y, pred_y = list(), list()
true_y, pred_y

i = 0
for train_index, test_index in loo.split(x_minmax):
    train_x, test_x = x_minmax.iloc[train_index, :], x2_minmax.iloc[test_index, :]
    i+=1
    
    train_y, test_y = y.iloc[train_index], y2.iloc[test_index]

    nan_values = test_x.isna()
    nan_columns = nan_values.any()
    
    columns_with_nan = test_x.columns[nan_columns].tolist()

    train_x = train_x.drop(columns_with_nan,axis = 1)
    test_x = test_x.dropna(axis='columns')

    #KNN Classifier
    classifier = KNeighborsClassifier(n_neighbors=3)
    classifier.fit(train_x, train_y)

    # evaluate model
    yhat = classifier.predict(test_x)
    yhatt=yhat.tolist()
    test_yy=test_y.tolist()

    # store
    true_y.append(test_yy[0])
    pred_y.append(yhatt[0])

acc = accuracy_score(true_y, pred_y)
acc = acc*100
error = 100-acc
print(pred_y)
print('Accuracy: %.3f' % acc)
print('Error: %.3f' % error)

print(classification_report(true_y, pred_y))
```

```python
#%%
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.neighbors import KNeighborsClassifier
from math import sqrt
from sklearn import model_selection
from sklearn.model_selection import LeaveOneOut
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import MinMaxScaler

data = pd.read_excel("hepatitis_new.xlsx", header=None)
data.drop(0, inplace=True, axis=1)
data.drop(0, inplace=True, axis=0)
data.columns = data.iloc[0] 
data.drop(1, inplace=True, axis=0)
data.columns = [c.replace(' ', '_') for c in data.columns]
# %%
data = data.replace(to_replace=['no', 'yes'], value=[0, 1])
data.CLASS = data.CLASS.replace(to_replace=['Live', 'Die'], value=[1, 0])
data = data.replace(to_replace=['?'], value=np.nan)
data = data.reset_index()
data = data.drop(columns=['index'])
X_temp = data.drop(columns=['CLASS']).values
y = data['CLASS'].values


#Fill NaN 
X = data.groupby("CLASS").transform(lambda x: x.fillna(x.mean()))
column = X.columns

#minmax
# min = 0
# max = 1
# minData = X.min()
# maxData = X.max()
# data_std = (X - X.min()) / (X.max()- X.min())
# X = data_std * (max - min) + min

std = X.std()
mean = X.mean()
#Zscore
x = (X-mean)/std

#Sigmoidal data train
X = (1-np.exp(-x))/(1+np.exp(-x))
X = X.values

#%%
accuracy_total=[]
error_total=[]
loo = LeaveOneOut()
neigh = KNeighborsClassifier(n_neighbors=3)

#scaler = StandardScaler()
for train_index, test_index in loo.split(X):
    X_train, X_test = X[train_index], X_temp[test_index]
    y_train, y_test = y[train_index], y[test_index]
    
    df_test = pd.DataFrame(X_test, columns = column)
    #Minmax
    # stdx = ((o - minData) / (maxData- minData))
    # X_test = stdx * (max - min) + min

    #Z score and Sigmoid (data test)
    x_temp = (df_test-mean)/std
    X_test = (1-np.exp(-x_temp))/(1+np.exp(-x_temp))
    X_test = X_test.values

    #Delete Atribut (NaN)
    t = np.argwhere(np.isnan(X_test))
    if len(t):
        for x_null in reversed(t):
            X_train = np.delete(X_train, x_null[1], 1)
    X_test = X_test[:, ~np.isnan(X_test).any(axis=0)]

    #knn
    neigh.fit(X_train, y_train)
    y_pred = neigh.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    error = 1 - accuracy
    accuracy_total.append(accuracy)
    error_total.append(error)

print("Acc: %.3f%%" % np.mean(accuracy_total))
print("Error: %.3f%%" % np.mean(error_total))



# %%
```
