```python
import pandas as pd
data = pd.read_csv("transaction.csv") 
data
```

```python
new_data = data[['Qty', 'InvoiceDate', 'Country']]
new_data['Month'] = pd.DatetimeIndex(new_data['InvoiceDate']).month
new_data['Year'] = pd.DatetimeIndex(new_data['InvoiceDate']).year
new_data = new_data.loc[(new_data['Country'] == 'Germany') & (new_data['Year'] == 2011)]
new_data = new_data.drop(columns=['InvoiceDate'])
new_data
```

```python
new_dataMonth = new_data['Month'].value_counts()
new_dataMonth = new_dataMonth.sort_index()
new_dataMonth
```

```python
import matplotlib.pyplot as plt

plt.plot(new_dataMonth)
plt.show()
```

# Linear Regression

```python
X = 0
Y = 0
XX = 0
XY = 0
print("X\t", "Y\t","XX\t", "XY")
for i in range (1,12+1):
    X = X + i
    Y = Y + new_dataMonth[i]
    XX = XX + i*i
    XY = XY + i*new_dataMonth[i]
    print(X, "\t", Y, "\t", XX, "\t",  XY)
print('='*50)

n=12
a = ((Y*XX) - (X*XY))/((n*XX) - X*X)
b = ((n)*(XY) - (X*Y))/((n*XX) - X*X)
print("a:", a)
print("b:", b)
print("==> Y =",a,"+",b,"*X")

# Predicted Quantity January 2012
Out = a + b*13
print('Predicted January 2012 = ', Out)
```
