```python
import pandas as pd
import numpy as np

data = pd.read_excel("hepatitis_new.xlsx", header=None)
data.drop(0, inplace=True, axis=1)
data.drop(0, inplace=True, axis=0)
data.columns = data.iloc[0]
data.drop(1, inplace=True, axis=0)
data.columns = [c.replace(' ', '_') for c in data.columns]
data = data.replace(to_replace=['no', 'yes'], value=[0, 1])
data.CLASS = data.CLASS.replace(to_replace=['Live', 'Die'], value=[0, 1])
data = data.replace(to_replace=['?'], value=np.nan)
data = data.reset_index()
X_temp = data.drop(columns=['CLASS'])
X_temp
```

```python
X = data.groupby("CLASS").transform(lambda x: x.fillna(x.mean()))
X
```

```python
y = data['CLASS'].values
y
```

```python
from sklearn import preprocessing
scaler = preprocessing.StandardScaler().fit(X)

X_scaled2 = scaler.transform(X)
X_scaled2
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=2, random_state=0).fit(X_scaled2)
kmeans.labels_
```

```python
# Accuracy
from sklearn.metrics import accuracy_score
acc = accuracy_score(y, kmeans.labels_)
acc*100
```

```python
kmeans_random = KMeans(init="random", n_clusters=2, random_state=0).fit(X_scaled2)
kmeans_random.labels_
```

```python
kmeans_random.cluster_centers_
```

```python
# Accuracy
from sklearn.metrics import accuracy_score
acc = accuracy_score(y, kmeans_random.labels_)
acc*100
```
