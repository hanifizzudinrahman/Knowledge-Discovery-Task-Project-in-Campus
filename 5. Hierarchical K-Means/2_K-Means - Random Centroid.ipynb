```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.datasets.samples_generator import make_blobs
from sklearn.cluster import KMeans
```

```python
X, y = make_blobs(n_samples=300, centers=2, cluster_std=0.60, random_state=0)
plt.scatter(X[:,0], X[:,1])
y
```

```python
kmeans = KMeans(n_clusters=2, init='random', random_state=0).fit(X)
kmeans.labels_
```

```python
for i in range (1,2):
    plt.scatter(X[:,0], X[:,1])
    kmeans = KMeans(n_clusters=2, init='random', random_state=i*5).fit(X)
    plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200, c='red')
    plt.show()
    print(kmeans.cluster_centers_)
    print(kmeans.labels_)
```

```python
# Accuracy
from sklearn.metrics import accuracy_score
acc = accuracy_score(y, kmeans.labels_)
acc*100        
```
