```python
import pandas as pd
import numpy as np
from sklearn.ensemble import IsolationForest

data={'Data 1': [80, 60, 75, 73, 81, 66], 
      'Data 2': [70, 84, 88, 90, 65, 60],
      'Data 3': [60, 40, 55, 58, 47, 49], 
      'Data 4': [60, 65, 60, 70, 68, 72],
      'Data 5': [60, 35, 40, 20, 56, 57], 
      'Data 6': [95, 98, 85, 87, 93, 95],
      'Data 7': [55, 70, 53, 64, 74, 77], 
      'Data 8': [50, 53, 57, 63, 58, 40],
      'Data 9': [62, 64, 53, 58, 40, 45]}

df=pd.DataFrame(data, columns=['Data 1','Data 2', 'Data 3','Data 4', 'Data 5','Data 6', 'Data 7','Data 8', 'Data 9'])
df
```

```python
# Transponse
df=df.T
df
```

```python
clf = IsolationForest(contamination=0.3)
pred = clf.fit_predict(df)

df['Outlier']=pred.reshape(-1,1)

df
```
