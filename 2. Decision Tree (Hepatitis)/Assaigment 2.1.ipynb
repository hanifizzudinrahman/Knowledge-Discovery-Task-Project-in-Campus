```python
import pandas as pd

# IMPORT DATASET
dataset = pd.read_csv('C:\Hanif Izzudin Rahman D4 EB 2016\S2 Elektro 2020 - PENS\Semester 2\Knowledge Discovery - Ali Ridho\Praktikum\P2_Decision Tree_Hepatitis\Assaigment2.1.csv')
dataset
```

```python
# Input Data
Input = dataset.iloc[:,0:19]
Input
```

```python
Input = Input.replace(to_replace=['no', 'yes'], value=[0, 1])
Input
```

```python
# Output Data
Output = dataset.iloc[:,-1]
Output
```

```python
from sklearn.model_selection import LeaveOneOut
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
from sklearn import tree
import numpy as np
```

```python
neighbor = DecisionTreeClassifier(criterion='gini',class_weight='balanced')
loo = LeaveOneOut()
Input = np.array(Input)
print(Input[:1])

Output = np.array(Output)
print(Output[:1])
acc = []
```

```python
for train_index, test_index in loo.split(Input,Output):
    X_train, X_test = Input[train_index], Input[test_index]
    y_train, y_test = Output[train_index], Output[test_index]
    neighbor.fit(X_train, y_train)
    y_pred = neighbor.predict(X_test)
    acc.append(accuracy_score(y_pred,y_test))

acc = np.array(acc)
sum_zero = np.count_nonzero(acc == 0)
print("Zero: ", sum_zero)

percent_error = (sum_zero/(len(acc))) * 100
print("Error: ", percent_error)
```
